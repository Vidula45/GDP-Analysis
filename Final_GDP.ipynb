### For the analysis below, use Data I-B. You can also use Data I-B along with Data I-A if required. Also, perform the analysis only for the duration 2014-15. 

```python
#Reading DATA-1B :
#creating a pathh for all the files.
path=r'..\Final_GDP\states_ut'
```

```python
#all the files are added into 'all_files'
all_files=glob.glob(path+'\*.csv')
all_files
```

```python
#Creating a dataframe from the all the files for each State and Union Territories.
list1=[]
for location in all_files:
    df=pd.read_csv(location,encoding = 'unicode_escape')
    df=df[['S.No.','Item','2014-15']]
    df=df.T
    df.columns=df.iloc[1]
    df=df[2:]
    df['State']=location.split('\\')[-1].split('.')[0].split('-')[1] #Creating a new column named 'STATE'. 
    df.head()
    list1.append(df) #adding all values to 'list1'
    
```

```python
#concating all the data from a list
df_new=pd.concat(list1,axis=0,ignore_index=True,sort=False)
df_new
```

```python
#Dropping all the sub-sub sectors as we need to deal only with Sectors and Sub-sectors.
df_new=df_new.drop('Crops',axis=1)
df_new=df_new.drop('Livestock',axis=1)
df_new=df_new.drop('Forestry and logging',axis=1)
df_new=df_new.drop('Fishing and aquaculture',axis=1)
df_new=df_new.drop('Trade & repair services',axis=1)
df_new=df_new.drop('Hotels & restaurants',axis=1)
df_new=df_new.drop('Railways',axis=1)
df_new=df_new.drop('Road transport',axis=1)
df_new=df_new.drop('Water transport',axis=1)
df_new=df_new.drop('Air transport',axis=1)
df_new=df_new.drop('Services incidental to transport',axis=1)
df_new=df_new.drop('Storage',axis=1)
df_new=df_new.drop('Communication & services related to broadcasting',axis=1)
df_new=df_new.drop('Services incidental to transport*',axis=1)
df_new=df_new.drop('Trade & repair services*',axis=1)
df_new=df_new.drop('Road transport**',axis=1)
df_new=df_new.drop('Road transport*',axis=1)

df_new
```

```python
#Reading Drop-out dataset:
dropout_df=pd.read_csv(r'..\Final_GDP\rs_session243_au570_1.1.csv')
dropout_df
```

```python
#Renaming the columns of dropout dataset:
dropout_df = dropout_df.rename(columns = {'Primary - 2014-2015':'Primary - 2013-2014','Primary - 2014-2015.1':'Primary - 2014-2015'})
```

```python
#Extracting the required columns from drop-out dataset:
#Level of Education, Primary-2014-2015,UpperPrimary-2014-15 and Secondary-2014-15
#we only require data for the year 2014-15.
dropout_df=dropout_df[['Level of Education - State','Primary - 2014-2015','Upper Primary - 2014-2015','Secondary - 2014-2015']]
```

###### Cleaning the drop-out dataset.

```python
#dropping the rows if there are any NAN values:
dropout_df=dropout_df.dropna(how='any')

#Renaming the columns as per the column names in DATA-1B.
#This is required for merging the 2 data sets.
dropout_df=dropout_df.rename(columns={'Level of Education - State':'State'}) #this is the common column in both the datasets hence renaming it to 'STATE'.
dropout_df=dropout_df.replace(['Andhra Pradesh'],['Andhra_Pradesh'])
dropout_df=dropout_df.replace(['Arunachal Pradesh'],['Arunachal_Pradesh'])
dropout_df=dropout_df.replace(['Chhatisgarh'],['Chhattisgarh'])
dropout_df=dropout_df.replace(['Himachal Pradesh'],['Himachal_Pradesh'])
dropout_df=dropout_df.replace(['Madhya Pradesh'],['Madhya_Pradesh'])
dropout_df=dropout_df.replace(['Tamil Nadu'],['Tamil_Nadu'])
dropout_df=dropout_df.replace(['Uttar Pradesh'],['Uttar_Pradesh'])
dropout_df=dropout_df.replace(['Uttrakhand'],['Uttarakhand'])

dropout_df
```

```python
#Merging the dropout-data with sector wise data(DATA-1B):

final_df=pd.merge(df_new,dropout_df,how='inner',on='State')
#Renaming the Population column for convenience.
final_df=final_df.rename(columns = {list(final_df)[19]:'Population'}) 
final_df.columns
```
